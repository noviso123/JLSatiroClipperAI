```python
# @title 1. Conectar Google Drive e Diagnóstico ☁️
import os
from google.colab import drive

print("🔍 DIAGNÓSTICO INICIAL...\n")

# 1. GPU Check
print("🎮 [1/2] Verificando Placa de Vídeo...")
exit_code = os.system('nvidia-smi > /dev/null 2>&1')
if exit_code == 0:
    print("    ✅ GPU NVIDIA Detectada! (Modo Turbo Ativado)")
else:
    print("    ⚠️ GPU NÃO DETECTADA! O sistema ficará lento.")
    print("    👉 Vá em 'Ambiente de Execução' -> 'Alterar tipo' -> 'T4 GPU' para corrigir.")

# 2. Drive Mount
print("\n☁️ [2/2] Conectando Google Drive (Force Remount)...")
if not os.path.exists('/content/drive'):
    try:
        drive.mount('/content/drive', force_remount=True)
    except Exception as e:
        print(f"    ⚠️ Erro Drive: {e}")
else:
    print("    ✅ Drive Já Conectado.")
```

```python
# @title 1.5. Verificação de Credenciais (Smart Auth) 🔑
import os
import shutil
from google.colab import files

TARGET_DIR = "/content/drive/MyDrive/JLSatiro_AI_Studio"
TARGET_FILE = os.path.join(TARGET_DIR, "client_secret.json")

print("🔑 VERIFICANDO CREDENCIAIS...")

if os.path.exists(TARGET_FILE):
    print(f"    ✅ Credenciais encontradas em: {TARGET_FILE}")
    print("    ⏩ Pulando etapa de upload (Já configurado!).")
else:
    print("    ⚠️ NENHUMA credencial encontrada.")
    print("    📂 Abrindo janela de upload... (Por favor, envie o 'client_secret.json')")

    try:
        uploaded = files.upload()
        for fn in uploaded.keys():
            if fn.endswith('.json'):
                if not os.path.exists(TARGET_DIR):
                    os.makedirs(TARGET_DIR, exist_ok=True)
                shutil.move(fn, TARGET_FILE)
                print(f"    💾 Salvo com segurança em: {TARGET_FILE}")
            else:
                print(f"    ❌ Arquivo '{fn}' ignorado (Não é JSON).")
    except Exception as e:
        print(f"    ❌ Erro no upload: {e}")
```
