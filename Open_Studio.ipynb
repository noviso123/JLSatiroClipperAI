```python
# @title 1. Conectar Google Drive e Diagn√≥stico ‚òÅÔ∏è
import os
from google.colab import drive

print("üîç DIAGN√ìSTICO INICIAL...\n")

# 1. GPU Check
print("üéÆ [1/2] Verificando Placa de V√≠deo...")
exit_code = os.system('nvidia-smi > /dev/null 2>&1')
if exit_code == 0:
    print("    ‚úÖ GPU NVIDIA Detectada! (Modo Turbo Ativado)")
else:
    print("    ‚ö†Ô∏è GPU N√ÉO DETECTADA! O sistema ficar√° lento.")
    print("    üëâ V√° em 'Ambiente de Execu√ß√£o' -> 'Alterar tipo' -> 'T4 GPU' para corrigir.")

# 2. Drive Mount
print("\n‚òÅÔ∏è [2/2] Conectando Google Drive (Force Remount)...")
if not os.path.exists('/content/drive'):
    try:
        drive.mount('/content/drive', force_remount=True)
    except Exception as e:
        print(f"    ‚ö†Ô∏è Erro Drive: {e}")
else:
    print("    ‚úÖ Drive J√° Conectado.")
```

```python
# @title 1.5. Verifica√ß√£o de Credenciais (Smart Auth) üîë
import os
import shutil
from google.colab import files

TARGET_DIR = "/content/drive/MyDrive/JLSatiro_AI_Studio"
TARGET_FILE = os.path.join(TARGET_DIR, "client_secret.json")

print("üîë VERIFICANDO CREDENCIAIS...")

if os.path.exists(TARGET_FILE):
    print(f"    ‚úÖ Credenciais encontradas em: {TARGET_FILE}")
    print("    ‚è© Pulando etapa de upload (J√° configurado!).")
else:
    print("    ‚ö†Ô∏è NENHUMA credencial encontrada.")
    print("    üìÇ Abrindo janela de upload... (Por favor, envie o 'client_secret.json')")

    try:
        uploaded = files.upload()
        for fn in uploaded.keys():
            if fn.endswith('.json'):
                if not os.path.exists(TARGET_DIR):
                    os.makedirs(TARGET_DIR, exist_ok=True)
                shutil.move(fn, TARGET_FILE)
                print(f"    üíæ Salvo com seguran√ßa em: {TARGET_FILE}")
            else:
                print(f"    ‚ùå Arquivo '{fn}' ignorado (N√£o √© JSON).")
    except Exception as e:
        print(f"    ‚ùå Erro no upload: {e}")
```
