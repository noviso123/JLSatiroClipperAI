```python
# @title 1. Conectar Google Drive e Diagn√≥stico ‚òÅÔ∏è
import os
from google.colab import drive

print("üîç DIAGN√ìSTICO INICIAL...\n")

# 1. GPU Check
print("üéÆ [1/2] Verificando Placa de V√≠deo...")
exit_code = os.system('nvidia-smi > /dev/null 2>&1')
if exit_code == 0:
    print("    ‚úÖ GPU NVIDIA Detectada! (Modo Turbo Ativado)")
else:
    print("    ‚ö†Ô∏è GPU N√ÉO DETECTADA! O sistema ficar√° lento.")
    print("    üëâ V√° em 'Ambiente de Execu√ß√£o' -> 'Alterar tipo' -> 'T4 GPU' para corrigir.")

# 2. Drive Mount
print("\n‚òÅÔ∏è [2/2] Conectando Google Drive (Force Remount)...")
if not os.path.exists('/content/drive'):
    try:
        drive.mount('/content/drive', force_remount=True)
    except Exception as e:
        print(f"    ‚ö†Ô∏è Erro Drive: {e}")
else:
    print("    ‚úÖ Drive J√° Conectado.")
```

```python
# @title 1.5. Upload de Credenciais (Recomendado) üîë
# Execute isso se quiser fazer login no YouTube com sua conta (Client Secret)
import os
import shutil
from google.colab import files

print("üîë UPLOAD DE CREDENCIAIS (CLIENT_SECRET.JSON)...")
print("    Fa√ßa upload do arquivo json baixado do Google Cloud Console.")
print("    Isso liberar√° acesso total ao Drive e YouTube API.\n")

try:
    uploaded = files.upload()

    for fn in uploaded.keys():
        print(f'    ‚úÖ Arquivo recebido: "{fn}"')
        if fn.endswith('.json'):
            # Move to persistent drive
            target_dir = "/content/drive/MyDrive/JLSatiro_AI_Studio"
            if not os.path.exists(target_dir):
                os.makedirs(target_dir, exist_ok=True)
            
            target_path = os.path.join(target_dir, "client_secret.json")
            shutil.move(fn, target_path)
            print(f"    üíæ Salvo com seguran√ßa em: {target_path}")
            print("    üöÄ ACESSO LIBERADO! Pode prosseguir para a Instala√ß√£o.")
        else:
            print("    ‚ö†Ô∏è Aviso: O arquivo n√£o parece ser um JSON. Ignorado.")
except Exception as e:
    print(f"    ‚ùå Erro no upload: {e}")
```
